**Amazon Product Review Analysis**

**Project Overview**  
In this project, I analyzed a dataset of Amazon product reviews. The goal is to explore how product categories, discounts, and other features relate to customer ratings. We'll use Python and visualizations to extract business insights.


```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean Seaborn style
sns.set(style='whitegrid')
```

##  Step 2: Load and Explore the Dataset
We start by loading the dataset and doing a basic inspection to understand the structure and quality of the data.


```python
# Load the dataset (make sure the file is uploaded in the Colab session)
df = pd.read_csv('amazon.csv')
df.head()
```

```python
# Check structure
df.info()

# Summary stats
df.describe()

# Check for missing values
df.isnull().sum()
```

# **Step 3: Data Cleaning**  
 We'll:
- Remove unwanted characters
- Convert columns to appropriate data types
- Drop rows with critical missing values


```python
# Strip whitespace from column names
df.columns = df.columns.str.strip()

# Drop rows with critical nulls (like missing product name, category, rating)
df.dropna(subset=['product_name', 'category', 'rating'], inplace=True)

# Fill others (like review_title) if you want to analyze reviews later
# df['review_title'] = df['review_title'].fillna("No Title")
```

```python
#  Check the first 10 unique values in 'discount_percentage' column
# Helps verify if values are strings with % or clean numeric
df['discount_percentage'].unique()[:10]
df['discount_percentage'].unique()[:10]
```

```python
# Remove % symbol and convert to float
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False).astype(float)
```

```python
# Create a flag column
df['high_discount'] = df['discount_percentage'] > 50
```

```python
df[['discount_percentage', 'high_discount']].head(10)
```

```python
# @title # Top 10 categories by number of reviews

top_categories = df['category'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title(" Top 10 Product Categories by Review Count")
plt.xlabel("Number of Reviews")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
#  Convert the 'rating' column from string type to numeric (float)
# This is required because originally the column might contain string values like "4.3"
# We need numeric values to perform calculations like average, correlation, etc.
# 'errors="coerce"' ensures any non-numeric or corrupted values are converted to NaN (safely ignored)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# @title # Average rating per category

avg_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating.values[:10], y=avg_rating.index[:10])
plt.title("Top 10 Categories by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

#  Discount vs Rating scatter plot


```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='discount_percentage', y='rating', alpha=0.4)
plt.title("💸 Discount % vs ⭐ Rating")
plt.xlabel("Discount Percentage")
plt.ylabel("Product Rating")
plt.tight_layout()
plt.show()
```

## 📈 Statistical Check: Correlation Between Discount and Rating


# 📈 Calculate the Pearson correlation between 'discount_percentage' and 'rating'
# This tells us whether there's a linear relationship between how much discount a product gets
# and how customers rate it on average
# A correlation near 0 → no relationship
# A correlation near 1 → strong positive relationship (higher discount = higher rating)
# A correlation near -1 → strong negative relationship (higher discount = lower rating)

```python
# Remove any stray '%' signs and convert to float safely
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
```

```python
# 📈 Measure the strength of the linear relationship between discount and rating
corr = df['discount_percentage'].corr(df['rating'])

# 🖨️ Display result nicely
print(f"Correlation between discount % and rating: {corr:.2f}")
```

## 💡 Insight: Discounts Don't Buy Happiness?

We found a weak negative correlation (-0.16) between the discount percentage and customer ratings.
This suggests that highly discounted products tend to receive slightly lower ratings — possibly due to:

- Lower perceived or actual quality
- Customers expecting more from discounted products
- More budget-focused products being rated more critically

This insight can help businesses rethink discount strategies: focus on **value and quality**, not just price drops.
